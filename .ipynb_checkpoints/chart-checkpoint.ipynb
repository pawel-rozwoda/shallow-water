```python
import matplotlib.pyplot as plt
import numpy as np
# xs = np.arange(0,10,1)
# plt.plot(xs,xs, '*')
# plt.text(2,7,'This text starts at point (2,4)',horizontalalignment='left')
axes = plt.gca()
axes.set_ylim([-0.25,1.7])
axes.set_xlim([-0.25,2.0])
data_x = [0,1,1.5]
data_y = [0, 1, 1.5]
plt.plot(data_x, data_y, '*')
plt.text(0,0,r' $v^{t-1}$',horizontalalignment='left', fontsize=20)
plt.text(1,1,r' $v^{t}$',horizontalalignment='left', fontsize=20)
plt.text(1.5,1.5,r' $v^{t+0.5}$',horizontalalignment='left', fontsize=20)
plt.savefig('v.png')
```
